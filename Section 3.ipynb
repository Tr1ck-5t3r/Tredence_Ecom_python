```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = orders_full[orders_full["order_status"] == "delivered"]
```

```python
region_groups = df.groupby("customer_state")["price"].apply(list)
```

```python
f_stat, p_value = stats.f_oneway(*region_groups)


print(f"F-statistic = {f_stat:.2f}")
print(f"P-value = {p_value:.4f}")

if p_value < 0.05:
    print("Statistically significant difference in average order values between regions.")
else:
    print("No significant difference in average order values between regions.")
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=orders_full, x="customer_state", y="order_value")
plt.xticks(rotation=90)
plt.title("Order Value Distribution by Region")
plt.xlabel("Customer State")
plt.ylabel("Order Value")
plt.tight_layout()
plt.show()
```

```python
# contingency table (payment_type vs order_status)
contingency_table = pd.crosstab(df["payment_type"], df["order_status"])
print(contingency_table)
```

```python
#Chi-Square Test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)


print(f"Chi2 = {chi2:.2f}, p-value = {p:.4f}")

if p < 0.05:
    print("Statistically significant relationship between payment method and order status.")
else:
    print("No significant relationship between payment method and order status.")
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=orders_full, x="payment_type", hue="order_status")
plt.title("Order Status by Payment Type")
plt.xlabel("Payment Type")
plt.ylabel("Number of Orders")
plt.legend(title="Order Status")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#correlation between price and freight_value
corr, p_val = stats.pearsonr(df["price"], df["freight_value"])
print("corr",corr)
print("p_val", p_val)


print(f"Correlation Coefficient = {corr:.2f}")
print(f"P-value = {p_val:.4f}")

if p_val < 0.05:
    print("Significant correlation between order value and freight charge.")
else:
    print("No significant correlation found.")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="order_value", y="freight_value", data=orders_full, alpha=0.6)
plt.title("Order Value vs Freight Value")
plt.xlabel("Order Value")
plt.ylabel("Freight (Shipping) Cost")
plt.tight_layout()
plt.show()
```

```python
category_data = df.dropna(subset=["category", "delivery_time"])

# Group delivery times by category
delivery_groups = category_data.groupby("category")["delivery_time"].apply(list)


#ANOVA test
f_stat, p_value = stats.f_oneway(*delivery_groups)


print(f"F-statistic = {f_stat:.2f}")
print(f"P-value = {p_value:.4f}")

if p_value < 0.05:
    print("Delivery time differs significantly across product categories.")
else:
    print("No significant difference in delivery time between categories.")
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x="category", y="delivery_time", data=orders_full)
plt.xticks(rotation=90)
plt.title("Delivery Time Distribution by Category")
plt.xlabel("Product Category")
plt.ylabel("Delivery Time (days)")
plt.tight_layout()
plt.show()
```
