## 4.4 Operational Metrics

#### Order status breakdown (delivered, canceled, returned, etc.)

```python
order_status_summary = df.groupby("order_status").agg(
    count=("order_status", "size"),
    percent=("order_status", lambda x: round(100 * len(x) / len(df), 2))
).reset_index()

print(order_status_summary)
```

```python
pd.crosstab(df["customer_state"], df["order_status"], normalize='index').T * 100
```

```python
status_counts = df.groupby(["product_id", "order_status"]).size().unstack(fill_value=0)
status_counts
```

```python
status_counts["total"] = status_counts.sum(axis=1)
status_counts["delivery_rate"] = status_counts["delivered"] / status_counts["total"]
status_counts["cancel_rate"] = status_counts["canceled"] / status_counts["total"]
```

```python
most_likely_delivered = status_counts.sort_values(by="delivery_rate", ascending=False).head(10)
least_likely_delivered = status_counts.sort_values(by="delivery_rate", ascending=True).head(10)

most_likely_cancelled = status_counts.sort_values(by="cancel_rate", ascending=False).head(10)
least_likely_cancelled = status_counts.sort_values(by="cancel_rate", ascending=True).head(10)
```

```python
print("Products most likely to be cancelled:")
most_likely_cancelled["cancel_rate"]
```

```python
print("Products most likely to be delivered:")
most_likely_delivered["delivery_rate"]
```

```python
print("Products least likely to be delivered:")
least_likely_delivered["delivery_rate"]
```

```python
print("Products least likely to be cancelled:")
least_likely_cancelled["cancel_rate"]
```

#### Average delivery time vs. estimated delivery by category

```python
df["estimated_delivery_time"] = (
    df["order_estimated_delivery_date"] - df["order_purchase_timestamp"]
).dt.days
delivered_df = df[df["order_status"] == "delivered"].copy()
delivered_df = delivered_df.dropna(subset=["delivery_time", "estimated_delivery_time"])
```

```python
category_delivery_comparison = delivered_df.groupby("category").agg(
    avg_actual_delivery_time=("delivery_time", "mean"),
    avg_estimated_delivery_time=("estimated_delivery_time", "mean")
).reset_index()
category_delivery_comparison
```

```python
bin_edges = list(range(0, 8000, 1000)) 
bin_labels = ["0-999", "1000-1999", "2000-2999", "3000-3999", "4000-4999", "5000-5999", "6000-7000"]

df["order_value_bin"] = pd.cut(df["order_value"], bins=bin_edges, labels=bin_labels, right=False)
```

```python
summary = df.groupby("order_value_bin").agg(
    avg_freight_cost=("freight_value", "mean"),
    avg_order_value=("order_value", "mean"),
    count=("order_id", "count")
).reset_index()
```

```python
plt.figure(figsize=(10,6))
plt.plot(summary["order_value_bin"], summary["avg_freight_cost"], marker='o', label='Avg Freight Cost')
plt.plot(summary["order_value_bin"], summary["avg_order_value"], marker='o', label='Avg Order Value')
plt.title("Freight Cost vs Order Value (1K Bins)")
plt.xlabel("Order Value Bin")
plt.ylabel("Amount")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Indicates freight value is not influenced by order value and it almost stays constant

## 4.5 Payment Analysis

#### Payment method distribution

```python
payment_counts = df["payment_type"].value_counts()
print(payment_counts)
payment_counts.plot(kind="bar",ylabel="counts")
```

#### Average payment value by payment type

```python
avg_payment_by_type = df.groupby("payment_type").agg(
    avg_payment_value=("payment_value", "mean"),
    count=("payment_value", "count")
).reset_index()
avg_payment_by_type = avg_payment_by_type.sort_values(by="avg_payment_value", ascending=False)
avg_payment_by_type
```

```python
plt.bar(avg_payment_by_type["payment_type"], avg_payment_by_type["avg_payment_value"])
plt.title("Average Payment Value by Payment Type")
plt.xlabel("Payment Type")
plt.ylabel("Average Payment Value")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

#### Payment method vs. order success/cancellation rates

```python
success_orders = df[df["order_status"] == "delivered"]
canceled_orders = df[df["order_status"] == "canceled"]
success_rate = success_orders.groupby("payment_type")["order_id"].count() / df.groupby("payment_type")["order_id"].count()
cancellation_rate = canceled_orders.groupby("payment_type")["order_id"].count() / df.groupby("payment_type")["order_id"].count()
```

```python
payment_rates = pd.DataFrame({
    "success_rate": success_rate,
    "cancellation_rate": cancellation_rate
}).reset_index()

# Fill missing values (in case there are payment types with no cancellations or successes)
payment_rates = payment_rates.fillna(0)
payment_rates["success_rate"] *= 100
payment_rates["cancellation_rate"]*= 100
payment_rates
```

```python
payment_rates.set_index("payment_type")[["success_rate", "cancellation_rate"]].plot(kind="bar", figsize=(10,6))
plt.title("Payment Method vs Order Success/Cancellation Rates")
plt.ylabel("Rate")
plt.xlabel("Payment Method")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
